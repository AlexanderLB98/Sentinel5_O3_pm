Para extraer información de ozono de los metadatos de la API de Sentinel-5, necesitas hacer lo siguiente:

Accede a la API de Sentinel-5: Primero, debes acceder a la API de Sentinel-5. Puedes hacerlo mediante una solicitud de API utilizando un lenguaje de programación como Python o mediante el uso de una herramienta como Postman.

Filtra por la variable de ozono: Una vez que tengas acceso a la API, deberás filtrar los datos para que solo se muestren los relacionados con la variable de ozono. Para hacerlo, deberás buscar en los metadatos de la API el nombre de la variable que deseas extraer. Por ejemplo, si deseas extraer datos de ozono, deberás buscar la variable "O3".

Selecciona la fecha y la ubicación: Una vez que hayas filtrado los datos para mostrar solo la variable de ozono, debes seleccionar la fecha y la ubicación de los datos que deseas extraer. Puedes hacerlo utilizando los parámetros de fecha y ubicación en la solicitud de API.

Extrae la información de ozono: Una vez que hayas seleccionado la fecha y la ubicación, podrás extraer la información de ozono de los datos de la API. Esta información puede estar en formato JSON o XML, dependiendo de la API que estés utilizando.

En resumen, para extraer información de ozono de los metadatos de la API de Sentinel-5, deberás acceder a la API, filtrar los datos para mostrar solo la variable de ozono, seleccionar la fecha y la ubicación, y extraer la información de ozono de los datos de la API.

```python
import requests
import json
```

```python
# URL de la API de Sentinel-5
url = "https://s5phub.copernicus.eu/dhus/api/stub/products"
```

```python

# Filtro de búsqueda para la variable de ozono
filter = {"filter": {"property": "Product_Level", "value": "L2__O3____"}}
```

```python
# Parámetros de fecha y ubicación
params = {
    "beginPosition": "2023-01-01T00:00:00.000Z",
    "endPosition": "2023-01-01T23:59:59.999Z",
    "bbox": "-180,-90,180,90",
}
```

```python
# Encabezados de solicitud
headers = {
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Content-Type": "application/json",
}
```

```python
auth = ('s5pguest', 's5pguest')
```

```python
# Solicitud de API
response = requests.post(url, headers=headers, params=params, json=filter, auth=auth)
```

```python
print(response.text)
```

```python
# Analizar la respuesta JSON
data = json.loads(response.text)
```

```python







# Extraer información de ozono
ozone_data = data["products"][0]["properties"]["O3_column_number_density"]

# Imprimir información de ozono
print(f"La densidad de columna de ozono es: {ozone_data}")
```

```python
from sentinel5dl import search, download
import json
```

```python


# Search for Sentinel-5 products
result = search(
        polygon='POLYGON((7.8 49.3,13.4 49.3,13.4 52.8,7.8 52.8,7.8 49.3))',
        begin_ts='2019-09-01T00:00:00.000Z',
        end_ts='2019-09-17T23:59:59.999Z',
        product='L2__CO____',
        processing_level='L2',
        processing_mode='Offline')

# Download found products to the local folder
download(result.get('products'), output_dir='data')
```

Ahora voy a poner la localización que tenemos:

```python
# loc = json.load('loc\madrid.geojson')
# Abrir el archivo en modo lectura
with open('loc\madrid.geojson', 'r') as archivo:
    # Leer el contenido del archivo
    contenido = archivo.read()

    # Convertir el contenido a un objeto Python
    datos = json.loads(contenido)

# Mostrar los datos
print(datos)
```

```python
# Obtener valores de "coordinates"
coordinates = []
for feature in datos['features']:
    coordinates.append(feature['geometry']['coordinates'])

print(coordinates)
```

```python
# Crear la cadena de salida
polygon = 'POLYGON((' + ','.join([f'{coordinates[0]} {coordinates[1]}' for coordinates in coordinates]) + '))'
print(polygon)
```

```python
# Usar el método join() para unir las coordenadas con comas y espacios
coord_string = ', '.join([f'{x} {y}' for x, y in coordinates])

# Concatenar la cadena resultante con el formato de la cadena de salida
result = f'POLYGON(({coord_string}))'
print(result)
```
